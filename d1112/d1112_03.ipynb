```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
x = [1,2,3]      
y = [2,4,8]
plt.plot(x,y)

plt.title("라인그래프 Line",size=30)   # 글자크기 변경
plt.xlabel('X축 날짜',color='red',loc='right')    # 그래프 하단 축 이름설정, 글자색, 글자위치 변경
plt.ylabel('Y축 인원',color='blue',loc='top')

plt.xticks([1,2,3])
plt.yticks([1,2,3,8,9])
plt.show()
```
