```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
x = [1,2,3]      
y = [2,4,8]
yy = [1,8,3]
plt.plot(x,y,label='인원변동',linewidth=3,marker='o',markersize=10,mec='blue')
# plt.plot(x,yy,label='비용변동',marker='v',ms=10,mfc='red',alpha=0.5,ls=':')
plt.plot(x,yy,'bv:',label='비용변동')   # 약식표현 'bv:' : 색상,마커,선모양

plt.title("라인그래프 Line",size=30)   # 글자크기 변경

## 범례 : 그래프 선 설명문구 loc=범례위치  upper,center,lower left,center,right
# plt.legend(loc='lower center')  
plt.legend(loc=(0.3,0.1))   # 좌표값으로 위치설정
plt.show()
```
