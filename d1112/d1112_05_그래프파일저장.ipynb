```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
x = df.index
y = df['국어']
z = df['영어']
e = df['수학']
plt.figure(figsize=(20,5),facecolor='yellow')   # x축으로 10, y축으로 10만큼 그래프가 커짐
plt.plot(x,y,label='국어',marker='v',mfc='yellow')
plt.plot(x,z,label='영어',marker='o',mfc='yellow')
plt.plot(x,e,label='수학',marker='x',mfc='yellow')
plt.legend()

# plt.savefig('graph2.png',dpi=300) # 그래프 이미지로 저장
```
