```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
## x,y축 데이터
x = [1,2,3]
y = [2,4,8]    # SM매출액
y2 = [5,1,3]   # LG매출액
y3 = [1,2,5]   # SK매출액
plt.figure(figsize=(10,5))
plt.plot(x,y,label='SM매출액',marker='o',ls='--')
plt.plot(x,y2,label='LG매출액',marker='x')
plt.plot(x,y3,label='SK매출액',marker='v',ls=':')

# for idx,txt in enumerate(y):
#   plt.text(x[idx],y[idx]+0.2,y[idx],ha='center')
for i in range(3):
  plt.text(x[i],y[i]+0.2,y[i],ha='center')
  plt.text(x[i],y2[i]+0.2,y2[i],ha='center')
  plt.text(x[i],y3[i]+0.2,y3[i],ha='center')

# plt.legend()   # 범례 세로출력(기본)
plt.legend(ncol=3)   # 범례 가로출력 ncol=n
plt.title("기업별 매출액")
plt.show()
```
