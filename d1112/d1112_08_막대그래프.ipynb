```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]
plt.figure(figsize=(3,4))
# plt.plot(label,values)
colors = ['r','g','b']
plt.title('막대그래프')
plt.bar(label,values,label='키',color=colors,width=0.5)
plt.xticks(label,['1번','2번','3번'],rotation=45)  # 라벨 변경, 기울기 설정
plt.yticks([0,50,100,150,200])
plt.legend(loc='upper center')
plt.show()
```

```python
colors = ['red','green','blue']
bar = plt.bar(label,values,width=0.5,color=colors)
# 막대그래프 무늬 설정
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')

plt.show()
```

```python
colors = ['red','green','blue']
bar = plt.bar(label,values,width=0.5,color=colors)

# 범위설정 xticks, yticks 항상 시작점 0부터 표시
plt.ylim(175,199)   # 원하는 시작점부터 표시가능 xlim, ylim
```

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]
colors = ['red','green','blue']
plt.title('막대그래프')
bar = plt.barh(label,values,height=0.5,color=colors)   # 옆으로 출력
plt.figure(figsize=(3,4))

plt.show()
```
