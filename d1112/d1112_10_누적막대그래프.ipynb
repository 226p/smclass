```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

누적 막대 그래프

```python
df = pd.read_excel('score.xlsx')
df
```

```python
plt.figure(figsize=(7,4))
plt.bar(df['이름'],df['국어'],label='국어')
# 누적막대 그래프 : bottom 첫번째 값을 입력
plt.bar(df['이름'],df['영어'],bottom=df['국어'],label='영어')
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],label='수학')
plt.xticks(rotation=45)
plt.legend(ncol=3,loc='upper center')
plt.ylim(0,350)

for i in range(8):
  plt.text(df['이름'][i],df['국어'][i]+5,df['국어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+10,df['영어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+df['수학'][i]+10,df['수학'][i],ha='center')

plt.show()
```
