```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
df = pd.read_excel('score.xlsx')
df
```

```python
plt.plot(df['이름'],df['키'])
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',linewidth=0.5)   # 안내선 긋기
plt.grid(axis='x',color='purple',alpha=0.3)   # 안내선 긋기

for i in range(8):
  plt.text(df['이름'][i],df['키'][i]+1,df['키'][i],ha='center')

plt.show()
```
