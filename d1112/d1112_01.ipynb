```python
df = pd.read_csv('drinks.csv')
df.shape
```

```python
df.info()
```

```python
# total_servings 컬럼을 구하시오 / 전체 알코올 도수 확인
df['total_servings'] = df['beer_servings']+df['spirit_servings']+df['wine_servings']
df
```

```python
df['continent'].unique()
df.groupby('continent')['total_servings'].sum()
```

```python
filt = df['continent'].isnull()
df.loc[filt,'total_servings']
```

```python
### null값을 'OT'로 채움
df['continent'] = df['continent'].fillna('OT')
```

```python
## 술 소비량 대비 알코올 비율 어느정도 되는지 출력
# 알코올 비율(total_litres_of_pure_alcohol/total_servings) 계산해서 컬럼(alcohol_rate) 추가
# 술소비량 대비 알코올 비율이 높은 나라 20개 출력
# 대륙별 알코올비율도 계산해서 출력
df.columns
```

```python
df['alcohol_rate'] = df['total_litres_of_pure_alcohol']/df['total_servings']
df[['country','alcohol_rate']].sort_values('alcohol_rate',ascending=False)
```

```python
df['alcohol_rate'].fillna(0,inplace=True)
df[['country','alcohol_rate']].sort_values('alcohol_rate',ascending=False).head(20)
```

```python
df[['country','alcohol_rate']].sort_values('alcohol_rate',ascending=False).head()
```
