```python
import pandas as pd
```

```python
# 상단 3개 삭제하고 가져오기
# index 행정기관 지정해서 df로 변환해서 출력
# usecols : 원하는 컬럼만 가져올 수 있음
df = pd.read_excel('연령별인구현황01.xlsx',skiprows=3,index_col='행정기관',usecols='B,E:Y')
df.head()
```

```python
# df[이름] : 컬럼을 보는 명령어      df['0~4세']
# df[슬라이싱] : row를 보는 명령어   df[1:3]
# df.iloc[번호] : row를 보는 명령어  df.iloc[3]
# df.loc[이름] : row를 보는 명령어   df.loc['0~4세']
df[0:1]
df.iloc[0]
df.loc['전국  ']
df.info()
# 데이터 replace함수 적용, 타입변경
df.iloc[0].str.replace(",","").astype(int)
```
