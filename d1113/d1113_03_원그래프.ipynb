```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
### 원그래프 pie
# values = [30,25,20,13,10,2]
values = [1]*6     # 100퍼센트 안 맞춰도 그 비율로 늘려줌
explode = [0.1,0,0,0,0,0]       # 원그래프 분리
labels = ['python','java','javascript','oracle','html','etc']   # 글자출력(시계반대 방향)
colors = ['y','g','r','g','b','c']
plt.title('프로그램 사용률')
# counterclock=False : 글자출력 방향 조정(시계방향)
# startangle=90 : 시작점 변경 12시부터 시작됨
# autopct='%.1f%%' :데이터값 출력 (소수점 첫째자리까지 % 붙여서 출력 / .0은 정수로 % 붙여서 출력)
plt.pie(values,labels=labels,explode=explode,autopct='%.0f%%')
plt.legend(loc=(1.2,0.3),title='언어별 선호도')

plt.show()
```
