```python
# 함수 사용하여 이름의 성 제거
# str.split() : 분할  /  str.startswith : 시작글자인식  / str.endswith() : 끝글자인식
# str.contains() : 포함글자 인식  /  str.find() : 문자위치 찾기  /  str.replace() : 문자대체
# str.pad() : 문자열 패딩, 특정글자로 채움  /  str.zfill() : 0으로 채움  /  str.strip() : 공백제거
# str.lower() : 소문자 변경  /  str.upper() : 대문자 변경  /  str.swapcase() : 대소문자 변경
def func(value):
  return value[1:]

df['이름'] = df['이름'].apply(func)
df
```

```python
# 요일 컬럼추가 -> 0~6까지 랜덤숫자 입력 -> 함수사용해서 0-일요일,1-월요일,2-화요일...
import random

df['요일']=0

for i in range(8):
  df.iloc[i,9]= random.randint(0,6)
```

```python
def func(value):
  if value==0: return'일요일'
  elif value==1: return '월요일'
  elif value==2: return '화요일'
  elif value==3: return '수요일'
  elif value==4: return '목요일'
  elif value==5: return '금요일'
  elif value==6: return '토요일'


df['요일'] = df['요일'].apply(func)
df
```
