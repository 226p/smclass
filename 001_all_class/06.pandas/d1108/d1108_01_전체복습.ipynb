```python
### 문자열 조건식 : startswith,endswith,ontains,isin

filt = df['이름'].str.startswith('김')
df[filt]
filt = df['이름'].str.endswith('김')
df.loc[filt,['이름','학교','키']]
filt = df['이름'].str.contains('수')
df.loc[filt,['이름','학교','키']]

# lower() : 해당값을 모두 소문자로 변환
# isin() : 포함되어 있는지 리스트만
lengs = ['Python','Java']
filt = df['SW특기'].str.lower().isin(lengs)
df[filt]

lengs = ['python','java']
filt = df['SW특기'].str.lower().isin(lengs)
df[filt]

### NaN데이터 처리하지 않으면 error
filt = df['SW특기'].str.contains('Java',na=False)
filt
```
