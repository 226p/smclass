```python
def func(x):
  if x-30<0 :  return x
  return x-30
```

```python
plt.figure(figsize=(7,4))
plt.title('학생성적그래프')
plt.bar(df['이름'],df['국어'],width=0.6,label='국어')
plt.bar(df['이름'],df['영어'],bottom=df['국어'],width=0.6,label='영어')
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],width=0.6,label='수학')
plt.legend()
n_list = df['국어'].apply(func)

for i in range(8):
  plt.text(df['이름'][i],df['국어'][i]+5,df['국어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+10,df['영어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+df['수학'][i]+10,df['수학'][i],ha='center')

plt.show()
```
