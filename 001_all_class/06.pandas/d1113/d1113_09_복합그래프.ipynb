```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import random
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
### 복합그래프 - 여러개 그래프
#  2x2개 그래프 혼합해서 사용
fig,axs = plt.subplots(2,2,figsize=(15,13))
# 메인타이틀 생성
fig.suptitle('4개 복합그래프')

# 4개 그래프 : 1번째 axs[0,0] / [0,1], [1,0], [1,1]
axs[0,0].bar(df['이름'],df['국어'],label='국어')    # 1번째 : 바차트
axs[0,0].set_title('1. 국어점수 그래프')
axs[0,0].set(xlabel='이름',ylabel='점수')  # x,y축 이름
axs[0,0].set_facecolor('lightyellow')    # 배경색
axs[0,0].legend(loc='upper center')  # 범례

axs[0,1].plot(df['이름'],df['영어'],label='영어')    # 2번째 : 선차트
axs[0,1].plot(df['이름'],df['수학'],label='수학')  
axs[0,1].set_title('2. 영어, 수학점수 그래프')
axs[0,1].set(xlabel='이름',ylabel='점수')  # x,y축 이름
axs[0,1].legend(loc='upper center')  # 범례

axs[1,0].barh(df['이름'],df['키'],label='키')    # 3번째 : 가로 바차트
axs[1,0].set_title('3. 키 가로 그래프')
axs[1,0].set_xlim(150,210)
axs[1,0].set(xlabel='키',ylabel='이름')  # x,y축 이름
axs[1,0].legend(loc='upper center')  # 범례
for i in range(8):
  axs[1,0].text(df['키'][i]+1,df['이름'][i],df['키'][i])

axs[1,1].plot(df['이름'],df['사회'],color='g',label='사회',alpha=0.5)    # 4번째 : 선차트
axs[1,1].set_title('4. 사회점수 그래프')
axs[1,1].set(xlabel='이름',ylabel='점수')  # x,y축 이름
axs[1,1].legend(loc='upper center')  # 범례
for i in range(8):
  axs[1,1].text(df['이름'][i],df['사회'][i]+2,df['사회'][i])







plt.show()
```
