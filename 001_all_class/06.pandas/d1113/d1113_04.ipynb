```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
# pct : values의 퍼센트 값 생성
# def custom_autopct(pct):
#   if pct >= 10:
#     # return ('%.1f%%' % pct)
#     # return ('{:.1f}%'.format(pct))
#     return (f'{pct:.1f}%')
#   else : return ''
  
def custom_autopct(pct):
  return f'{pct:.1f}%' if pct >= 10 else ''
```

```python
## 원그래프
values = [30,25,20,13,10,2]  # 100퍼센트
# values = [1,1,1,1,2,1]  # 100퍼센트
explode = [0.0,0,0,0,0,0]
wedgeprops = {'width':0.6,'edgecolor':'white','linewidth':2}   # 중간공백(도넛모양)
labels = ['python','java','javascript','oracle','html','etc']
colors = ['plum','beige','skyblue','bisque','khaki','peachpuff']
plt.title('프로그램 사용률')
# pctdistance=0.7 : 글자위치 설정
plt.pie(values,labels=labels,explode=explode,autopct=custom_autopct,colors=colors,wedgeprops=wedgeprops,\
        startangle=90,counterclock=False,pctdistance=0.7)
plt.legend(loc=(1.2,0.3),title='언어별 선호도')
# plt.legend(loc='upper right')
plt.show()
```
