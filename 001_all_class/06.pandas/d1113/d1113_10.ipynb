```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import random
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
df = pd.read_csv('chipotle.tsv',sep='\t')
df.tail()
```

```python
df.columns
```

```python
### 메뉴별 주문개수 / 순서 상관없이 5개 막대그래프로 출력
menu = df.groupby('item_name')['quantity'].sum().head(10)
avg = df.groupby('item_name')['quantity'].sum().mean()
avg
menu['mean']=avg
```

```python
plt.figure(figsize=(5,3))
list = plt.bar(menu.index,menu)
x = menu.index
y = [avg]*11
list[len(menu.index)-1].set_color('blue')
plt.plot(x,y,color='red',ls='--')
plt.grid(axis='y',alpha=0.5)
plt.xticks(rotation=90)
for i in range(11):
  plt.text(menu.index[i],menu[i]+1,menu[i],ha='center')

plt.show()
```
