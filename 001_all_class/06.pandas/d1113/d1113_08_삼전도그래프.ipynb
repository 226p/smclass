```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import random
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
### 삼전도그래프
df['학년'] = [3,3,2,1,1,3,2,2]
```

```python
# size = np.random.rand(8)*1000  # 원의 크기 랜덤
size = df['학년']*500
plt.title('산점도 그래프')
plt.xlabel('영어')
plt.ylabel('수학')
# s : 산점도 크기 변경 / c = 산점도 표시컬럼
# cmap : matplotlib 색상표 사용 하 컬럼내용 색상 선택
plt.scatter(df['영어'],df['수학'],s=size,c=df['학년'],cmap='plasma',alpha=0.5)
# colorbar : 색상바 생성 - 산점도색상의 학년표시
# ticks : 바범위
# shrink : 바크기 / orientation : 'horizontal' 바 하단배치, 옆배치 선택
plt.colorbar(ticks=[1,2,3],label='학년',shrink=1)

plt.show()
```
