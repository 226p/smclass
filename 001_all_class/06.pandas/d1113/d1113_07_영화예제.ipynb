```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
data = df['관객 수'].mean()
```

```python
movie = df.groupby('영화')['관객 수'].mean()
```

```python
movie.index
```

```python
df.index[0]
```

```python
movie['평균'] = data
```

```python
data
```

```python
plt.figure(figsize=(15,10))
s_bar = plt.bar(movie.index,movie)
# plt.xticks(rotation=90)
x = movie.index
y = [data]*9
s_bar[len(movie.index)-1].set_color('blue')
plt.ylim(1000,1800)  # 시작점을 1000으로 1800까지 범위를 줌
plt.plot(x,y,color='red',ls='--')

plt.show()
```
