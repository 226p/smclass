```python
import pandas as pd
import numpy as np
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
# index 정렬, ascending=False 역순정렬
df.sort_index(ascending=False)
df

# sort_values : 컬럼의 값으로 정렬
df.sort_values('키',ascending=False)
df.sort_values('이름')
```

```python
# 기준점 국어로 영어점수 정렬
df.sort_values(['국어','영어'],ascending=False)
df
```

```python
df.sort_values(['국어','영어'],ascending=[False,True])
df
```
