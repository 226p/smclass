```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False 
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
```

```python
df = pd.read_csv('drinks.csv')
df
```

```python
df['total_servings']=df['beer_servings']+df['spirit_servings']+df['wine_servings']
df['total_servings'] = df['total_servings'].dropna()
```

```python
df.shape
```

```python
filt = df['country'].str.lower().str.contains('korea')
df[filt]
```

```python
index = np.arange(df.shape[0])
plt.bar(index,df['total_servings'])
# plt.xticks(index)
for i in range(198):
  plt.text(i,df['total_servings'][i],df['total_servings'][i])
  # if df['country'].str.lower().str.contains('south_korea'):
  #   color='red'
plt.title('나라별 그래프')

plt.show()
```
